```python
def df_formating(path, donor=None):
    
    index = {"Quantity": ["Qty", "Quantity"], 
             "Reference": ["Book No.", "PATNumber", "Reference"], 
             "Serial Number": ["Serial no.", "Serial number", "Serialnumber asset"], 
             "Asset Tag": ["Asset Tag", "Client Asset Tag"], 
             "Type": ["Product type", "Product"], 
             "Brand": ["Brand", "Manufacturer"], 
             "Model": ["Model"], 
             "Processor": ["Processor"], 
             "RAM": ["Memory", "RAM"], 
             "HDD": ["Drive Size", "HDD"], 
             "Display": ["DISPLAY SIZES", "Screen size"], 
             "Weight": ["Weight (KG)", "Weight"], 
             "Grade": ["Grade", "Condition Code", "GRADE-IN"], 
             "Defects": ["Defects", "Stock Comments"],
             "Disk Status": ["Disk Status", "Erasure Status"],
             "Donor": ["Donor"]
            }
    
    excel_file = path.split("/")

    if excel_file[-1][:2] == "LB":
        df = pd.read_excel(path, header=[2])
        new = df["Specifications"].str.split(",", expand=True)[0][:2]
        df["Processor"] = str(new[0]) + "," + str(new[1])
        df["RAM"] = df["Specifications"].str.split(",", expand=True)[2]
        df["HDD"] = df["Specifications"].str.split(",", expand=True)[3]
        df["Reference"] = df["Load no."].astype(str) + " - " + df["Tracking reference"].astype(str)
        
        if donor is None:
            try:
                df["Donor"] = donor            
            except:       
                df["Donor"] = pd.read_excel(path, usecols="F", header=None, names=["Donor"]).iloc[0]["Donor"]        
        
        
    elif excel_file[-1][:2] == "CT":
        df = pd.read_excel(path)
        df["Defects"] = df["Defect 1"] + "," + df["Defect 2"] + "," + df["Defect 3"] + "," + df["Defect 4"] + "," + df["Defect 5"]
        donor_name = path.split(" ")
        del(donor_name[:3], donor_name[-2:])
        df["Donor"] = " ".join(donor_name)
        
        if donor:
            try:
                df["Donor"] = donor            
            except:       
                df["Donor"] = pd.read_excel(path, usecols="F", header=None, names=["Donor"]).iloc[0]["Donor"]
                 
    else:
        df = pd.read_excel(path, header=[18])
        df["Donor"] = pd.read_excel(path, usecols="B", header=None, names=["Donor"]).iloc[2]["Donor"]
    
    columns = list(df)
    new_df = pd.DataFrame(columns=index)
    print(list(new_df))
    
    print(df.head())
    
    for k, v in index.items():
        print("k", k)
        print("v", v)
        for i in v:
            try:
                new_df[k] = df[i]
            except:
                continue             
                
    new_df["Quantity"] = new_df["Quantity"].fillna(1)
    new_df["Type"] = new_df["Type"].fillna("PC")
            
        
    return new_df
```

```python
class Df_formating():
    
    
    def __init__(self, path, donor=None):
        
        self.path: str = path
        self.donor: str = donor
        self.excel_file = path.split("/")
        self.index: dict = {"Quantity": ["Qty", "Quantity"], 
             "Reference": ["Book No.", "PATNumber", "Reference"], 
             "Serial Number": ["Serial no.", "Serial number", "Serialnumber asset"], 
             "Asset Tag": ["Asset Tag", "Client Asset Tag"], 
             "Type": ["Product type", "Product", "Stock Type"], 
             "Brand": ["Brand", "Manufacturer"], 
             "Model": ["Model"], 
             "Processor": ["Processor"], 
             "RAM": ["Memory", "RAM"], 
             "HDD": ["Drive Size", "HDD"], 
             "Display": ["DISPLAY SIZES", "Screen size"], 
             "Weight": ["Weight (KG)", "Weight"], 
             "Grade": ["Grade", "Condition Code", "GRADE-IN"], 
             "Defects": ["Defects", "Stock Comments"],
             "Disk Status": ["Disk Status", "Erasure Status"],
             "Donor": ["Donor"]
            }
        
        if self.excel_file[-1][:2] == "LB":
            self.lb_parsing()
        elif self.excel_file[-1][:2] == "CT":
            self.ct_parsing()
        else:
            self.tes_parsing()
            
        
    def lb_parsing(self):
        
        self.df = pd.read_excel(self.path, header=[2])
        new = df["Specifications"].str.split(",", expand=True)[0][:2]
        self.df["Processor"] = str(new[0]) + "," + str(new[1])
        self.df["RAM"] = self.df["Specifications"].str.split(",", expand=True)[2]
        self.df["HDD"] = self.df["Specifications"].str.split(",", expand=True)[3]
        self.df["Reference"] = self.df["Load no."].astype(str) + " - " + self.df["Tracking reference"].astype(str)
        return self.create_df(self.df)
        
        
    def ct_parsing(self):
        
        self.df = pd.read_excel(self.path)
        self.df["Defects"] = self.df["Defect 1"] + "," + self.df["Defect 2"] + "," + self.df["Defect 3"] + "," + self.df["Defect 4"] + "," + self.df["Defect 5"]
        donor_name = self.path.split(" ")
        del(donor_name[:3], donor_name[-2:])
        self.df["Donor"] = " ".join(donor_name)
        return self.create_df(self.df)
        
        
    def tes_parsing(self):
        
        #Create a dictionary to map the types to match CO2 calculator 
        self.types = {"LT": "Laptop", "PC": "Desktop", "LD": "Monitor", "": "Others"}
        self.df = pd.read_excel(self.path, header=[18])
        self.df["Donor"] = pd.read_excel(self.path, usecols="B", header=None, names=["Donor"]).iloc[2]["Donor"]
        return self.create_df(self.df)
    
        
    def create_df(self, df):
        
        self.df = df
        print(self.df)
        self.new_df = pd.DataFrame(columns=self.index)
        for k, v in self.index.items():
            for i in v:
                try:
                    self.new_df[k] = self.df[i]
                except:
                    continue
                    
        if self.donor != None:
            self.new_df["Donor"] = self.donor
            
        self.new_df = self.new_df.replace("-", np.nan)
            
        return self.new_df
```
